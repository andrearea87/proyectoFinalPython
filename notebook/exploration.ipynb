```python
import pandas as pd
import matplotlib.pyplot as plt

#Generar graficos estadisticos

#Sitio web 1
print("ESTADISTICAS SITIO WEB 1")
data_path_1="../data/productos_web_1.csv"
df=pd.read_csv(data_path_1)
#Barras
plt.hist(df["Precio"], bins=20,edgecolor="black")
plt.title("Distribucion de precios")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
#Pastel
y=0
precios=[]
etiquetas=[]
for x in range(8):
    cantidad=len(df[(df["Precio"]>=y) & (df["Precio"]<(y+50))])
    if(cantidad>10):
        precios.append(cantidad)   
        etiquetas.append(str(cantidad)+" productos en el rango "+str(y)+"-"+str(y+49)+" USD")
    y+=50
plt.pie(precios, labels = etiquetas)
plt.title("Distribucion de precios")
plt.show() 

#Sitio web 2
print("ESTADISTICAS SITIO WEB 2")
data_path_2="../data/productos_web_2.csv"
df2=pd.read_csv(data_path_2)
#Barras
plt.hist(df2["Precio"], bins=20,edgecolor="black")
plt.title("Distribucion de precios")
plt.xlabel("Precio (USD)")
plt.ylabel("Frecuencia")
plt.show()
#Pastel
j=0
precios2=[]
etiquetas2=[]
for x in range(8):
    cantidad=len(df2[(df2["Precio"]>=j) & (df2["Precio"]<(j+50))])
    if(cantidad>10):
        precios2.append(cantidad)   
        etiquetas2.append(str(cantidad)+" productos en el rango "+str(j)+"-"+str(j+49)+" USD")
    j+=50
plt.pie(precios2, labels = etiquetas2)
plt.title("Distribucion de precios")
plt.show() 



```
